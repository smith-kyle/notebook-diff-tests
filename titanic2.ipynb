This notebook will explore the Titanic dataset, focusing on the relationship between fare prices and survival rates among the passengers. We will create visualizations to better understand how the fare prices paid by passengers might have impacted their chances of survival.

Dependencies:
- seaborn
- matplotlib
- pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cell 3 - Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Cell 4 - Data preprocessing
# Removing rows with missing 'fare' values
titanic = titanic.dropna(subset=['fare'])

# Cell 5 - Fare distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic, x='fare', kde=True, bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare Price')
plt.ylabel('Frequency')
plt.show()
```

The histogram above shows the distribution of fare prices paid by the passengers aboard the Titanic. The distribution is right-skewed, with most passengers paying lower fares and a few passengers paying significantly higher fares.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x='survived', y='fare', hue='class', palette='Paired')
plt.title('Fare vs. Survival by Class')
plt.xlabel('Survival Status')
plt.ylabel('Fare Price')
plt.legend(title='Class')
plt.show()
```

The boxplot above compares the fare prices paid by survivors and non-survivors, separated by class. We can observe the following trends:
- Passengers who survived generally paid higher fares than those who did not survive.
- The median fare price for survivors was higher than that of non-survivors in all three classes.

This suggests that passengers who paid higher fares had a greater chance of survival, possibly due to the better accommodations and closer proximity to lifeboats in higher classes.
